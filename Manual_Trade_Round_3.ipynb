```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def PNL_1(Bid):
    a= 160
    b = 200
    c = 250
    d = 320
    if Bid <=b:
        return((d-Bid)*(Bid-a))
    if Bid >b and Bid <=c:
        return((d-Bid)*(b-a))
    if Bid >c:
        return((d-Bid)*(Bid-c))+((d-Bid)*(b-a))
```

```python
Bids = np.linspace(160, 320, 1000)
PNLs = [PNL_1(b)/110 for b in Bids]
plt.figure(figsize=(10, 6))
plt.plot(Bids, PNLs, label='PNL Curve', color='blue')
plt.title('PNL Curve')
plt.xlabel('Bid Price')
plt.ylabel('PNL')
plt.legend()
plt.grid()
plt.show()
```

```python
def PNL_2(Bid):
    return((320-Bid)*((Bid-250)/70))
```

```python
Bids = np.linspace(250, 320, 100)
PNLs = PNL_2(Bids)
plt.figure(figsize=(10, 6))
plt.plot(Bids, PNLs, label='PNL Curve', color='blue')
plt.title('PNL Curve')
plt.xlabel('Bid Price')
plt.ylabel('PNL')
plt.legend()
plt.grid()
plt.show()
```

```python
def penalty(bid,average):
    return (((320-average)/(320-bid))**3)
```

```python
average = np.linspace(285, 320, 1000)
penalties = [penalty(285, avg) for avg in average]
plt.figure(figsize=(10, 6))
plt.plot(average, penalties, label='Penalty Curve', color='blue')
plt.title('Penalty Curve')
plt.xlabel('Average Price')
plt.ylabel('Penalty')
plt.legend()
plt.grid()
plt.show()
```
